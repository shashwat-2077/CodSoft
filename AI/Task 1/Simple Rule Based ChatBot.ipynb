# Finding the Best Matching Pattern

I've implemented a function to identify the best matching pattern from the 'rules' dictionary based on a similarity ratio. This ensures that the chatbot provides relevant responses.

Defining a function that finds the best matching pattern from the rules dictionary based on similarity ratio.

Args: user_input (str): The user's input query.

Returns: tuple: A tuple containing the best matching pattern and its associated similarity ratio.

```python
def find_best_match(user_input):
    best_match = None
    highest_ratio = 0
    for pattern in rules.keys():
        ratio = fuzz.ratio(user_input.lower(), pattern)
        if ratio > highest_ratio:
            highest_ratio = ratio
            best_match = pattern
    return best_match, highest_ratio
```

# Auto Spelling Correction
To enhance user input processing, the ChatBot corrects spelling mistakes using the 'textblob' library, making it more user-friendly and accurate.


```python
def correct_spelling(input_text):
    blob = TextBlob(input_text)
    corrected_text = blob.correct()  # Correct spelling mistakes
    return str(corrected_text)
```

# Feedback Mechanism
After each interaction, the chatbot prompts users to provide feedback on the response's helpfulness. Users can also suggest improvements, allowing us to continuously enhance the chatbot.


```python
positive_feedback_log = []
negative_feedback_log = []

def collect_feedback(response, user_input):
    feedback = input("Did my response help you? (Yes/No): ").strip().lower()
    if feedback == "yes" or feedback == "y":
        positive_feedback_log.append({"user_input": user_input, "response": response})
        print("Thank you for your feedback! I'm glad I could help.")
    elif feedback == "no" or feedback == "n":
        improvement_suggestion = input("What could I have done better? Please provide suggestions: ")
        negative_feedback_log.append({"user_input": user_input, "response": response, "suggestion": improvement_suggestion})
        print("Thank you for your feedback! I'll use your suggestions to improve.")
    else:
        print("Thank you for your feedback!")
```

```python
def respond(user_input):

    # Correct user input's spelling
    corrected_input = correct_spelling(user_input)

    # Find the best matching pattern and its similarity ratio
    best_match, ratio = find_best_match(user_input)

    # Check if a suitable match was found and the similarity ratio is above the threshold
    if best_match and ratio >= 70:  # Adjust the similarity threshold as needed
        response = rules[best_match]
        return response
    else:
        default_response = rules[".*"]
        return default_response
```

```python
def main():
    # Display initial greeting
    print("ChatBot: Hello! How can I assist you today?")

    # Initialize conversation history as an empty list
    conversation_history = []

    # Start an infinite loop to keep the conversation going
    while True:
      try:
        # Prompt the user for input
        user_input = input("You: ")

        # Check if the user wants to exit the conversation
        if user_input.lower() == "exit":
            print("ChatBot: Goodbye! Have a great day!")
            break

        # Store the user input in the conversation history
        conversation_history.append(("You", user_input))

        # Get a response from the chatbot based on user input
        response = respond(user_input)

        # Store the chatbot's response in the conversation history
        conversation_history.append(("ChatBot", response))

        # Display the chatbot's response
        print("ChatBot:", response)

        # Display the conversation history so far
        print("\nConversation History:")
        for role, message in conversation_history:
            print(f"{role}: {message}")

        # Collect feedback after displaying the response
        collect_feedback(response, user_input)

      # Handle exceptions (unexpected errors)
      except Exception as e:
            print("ChatBot: An error occurred. Please try again or type 'exit' to end the conversation.")
            print("Error:", e)


if __name__ == "__main__":
    # Call the main function when the script is run directly
    main()

# After the chatbot interaction loop, you can analyze and view the feedback logs
print("\nPositive Feedback Log:")
for feedback in positive_feedback_log:
    print("User Input:", feedback["user_input"])
    print("Response:", feedback["response"])
    print()

print("\nNegative Feedback Log:")
for feedback in negative_feedback_log:
    print("User Input:", feedback["user_input"])
    print("Response:", feedback["response"])
    print("Suggestion:", feedback["suggestion"])
    print()
```
